```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the file
file_path = "C:/Users/ramon/Documents/MASTER BIG DATA & DATA SCIENCE/TFM/Nueva carpeta/TFM-main/datas/Option_B/Brazil_Energy balance matrix.xlsx"

#File excel with multi-sheet_names is read as an Dictionary
Dict = pd.read_excel(file_path, sheet_name=None, skiprows=range(4), skipfooter=3)

for k in Dict.keys():       
    # Round the values of each column in dataframe with 2 decimals
    Dict[k]=Dict[k].applymap(lambda x: round(x, 2) if isinstance(x, (int, float)) else x)
    
    #Ignor the first row (unit) in each df
    Dict[k]=Dict[k][1:]
    
    #Rename of columns in df
    Dict[k].rename(columns={'Unnamed: 0': 'SECTOR','OTHER PRIMARY_x000d_\n':'OTHER PRIMARY'}, inplace=True)   
    
print(Dict.keys())

# Convert the keys of Dict to only contain the year
for k in list(Dict.keys()):
    k_new = k.split(' - ')[0]
    Dict[k_new] = Dict.pop(k)
    
    Dict[k_new]["Year"]=k_new
        
print('\n')
print(Dict.keys())



#Save the Clean Dictionary in Excel with multi-sheet_names
file_path="C:/Users/ramon/Documents/MASTER BIG DATA & DATA SCIENCE/TFM/Nueva carpeta/TFM-main/datas/Option_B/Brazil_Energy balance matrix_cleaned.xlsx"
with pd.ExcelWriter(file_path) as writer:
    for k, df in Dict.items():
        df.to_excel(writer, sheet_name=k)

df_2021=Dict['2021']
df_2021.tail(14)
```

```python
Dict2=Dict.copy()

for k in Dict2.keys():
   # selected_columns = ['SECTOR', 'OIL', 'NATURAL GAS', 'COAL', 'HYDROENERGY', 'GEOTHERMAL',
    #   'NUCLEAR', 'FIREWOOD', 'SUGARCANE AND PRODUCTS', 'OTHER PRIMARY', 'TOTAL PRIMARIES','Year']
    selected_columns = ['SECTOR','TOTAL','Year']
    Dict2[k]=Dict2[k][selected_columns]
    selected_rows=['   TRANSPORT ','FINAL CONSUMPTION']
    Dict2[k]=Dict2[k][Dict2[k]['SECTOR'].isin(selected_rows)]
    
    # Concatenate all DataFrames in Dict vertically
    df_conc = pd.concat(Dict2.values(), ignore_index=True)
    
    # Move the column "Year" to the first position
    last_column_name = df_conc.columns[-1]
    # Reorder the columns by concatenating the last column with the rest of the DataFrame
    df_conc = pd.concat([df_conc[last_column_name], df_conc.drop(columns=last_column_name)], axis=1) # Datafames of TOTAL SUPPLY
    df_conc.set_index("Year", inplace=True)
    df_conc.index = df_conc.index.astype(int)
    #df_TC1=df_conc.drop(columns="SECTOR")
    # Set the value "NaN" by Zero
    df_TC1 = df_conc.apply(lambda column: column.fillna(0))
    

df_TC1
```

```python
# Group by Year and SECTOR, and sum the TOTAL values
grouped_df = df_TC1.groupby(['Year', 'SECTOR'])['TOTAL'].sum().reset_index()

# Pivot the data
pivot_df = grouped_df.pivot(index='Year', columns='SECTOR', values='TOTAL')

# Display the result
pivot_df.head()
```

```python
pivot_df.columns = pivot_df.columns.str.strip()
```

```python
pivot_df['TRANSPORT/FC'] = pivot_df['TRANSPORT'] / pivot_df['FINAL CONSUMPTION']
```

```python
pivot_df['TRANSPORT/FC'].fillna(0, inplace=True)
```

```python
pivot_df
```

```python
media = pivot_df['TRANSPORT/FC'].mean()
desviacion_tipica = pivot_df['TRANSPORT/FC'].std()
moda = pivot_df['TRANSPORT/FC'].mode().iloc[0] # La moda puede tener más de un valor, así que tomamos el primero
mediana = pivot_df['TRANSPORT/FC'].median()

print("La media de TRANSPORT/FC es:", media)
print("La desviación típica de TRANSPORT/FC es:", desviacion_tipica)
print("La moda de TRANSPORT/FC es:", moda)
print("La mediana de TRANSPORT/FC es:", mediana)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Suponiendo que tus datos estén en orden cronológico
X = np.arange(len(pivot_df)).reshape(-1, 1)
y = pivot_df['TRANSPORT/FC'].values

model = LinearRegression()
model.fit(X, y)

# Prediciendo 20 puntos en el futuro
X_future = np.arange(len(pivot_df), len(pivot_df) + 20).reshape(-1, 1)
predictions = model.predict(X_future)

print("Predicciones para los próximos 20 años:", predictions)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Suponiendo que tus datos estén en orden cronológico
X = np.arange(len(pivot_df)).reshape(-1, 1)
y = pivot_df['TRANSPORT/FC'].values

model = LinearRegression()
model.fit(X, y)

# Prediciendo 20 puntos en el futuro
X_future = np.arange(len(pivot_df), len(pivot_df) + 20).reshape(-1, 1)
predictions = model.predict(X_future)

# Gráfico de los datos actuales
plt.plot(X, y, label="Datos Actuales")

# Gráfico de las predicciones
plt.plot(X_future, predictions, label="Predicciones a 20 años", linestyle='dashed')

plt.xlabel("Tiempo")
plt.ylabel("TRANSPORT/FC")
plt.legend()
plt.title("Predicción a 20 años de TRANSPORT/FC")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Asumiendo que tus datos están en orden cronológico y son estacionarios
y = pivot_df['TRANSPORT/FC'].values

# Ajustando el modelo ARIMA
# Puedes ajustar los parámetros p, d, q según tu serie temporal
model = ARIMA(y, order=(5,1,0))
fit_model = model.fit()

# Haciendo una predicción para los próximos 20 puntos
forecast = fit_model.forecast(steps=20)

# Gráfico de los datos actuales
plt.plot(y, label="Datos Actuales")

# Gráfico de las predicciones
plt.plot(range(len(y), len(y) + 20), forecast, label="Predicciones a 20 años", linestyle='dashed')

plt.xlabel("Tiempo")
plt.ylabel("TRANSPORT/FC")
plt.legend()
plt.title("Predicción a 20 años de INDUSTRIAL/FC con ARIMA")
plt.show()
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# Kernel para el GP
kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))

# Crear y entrenar el modelo
gp_model = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
gp_model.fit(X, y)

# Hacer predicciones a 20 años
predictions, sigma = gp_model.predict(X_future, return_std=True)

plt.plot(y, label="Datos Actuales")
plt.plot(range(len(y), len(y) + 20), predictions, label="Predicciones a 20 años", linestyle='dashed')
plt.fill_between(range(len(y), len(y) + 20), predictions - sigma, predictions + sigma, alpha=0.2)
plt.legend()
plt.show()
```
